```python
# Magic comment
%matplotlib inline

# Import libraries
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
```

```python
# Load filles
N_rep = 12
df_index = pd.read_table('replica_index.xvg', delim_whitespace=True,  skiprows=1, names=['%d' % i for i in range(0, N_rep)]).reset_index().drop(['index'], axis=1)
df_tindex = pd.read_table('replica_temp.xvg', delim_whitespace=True,  skiprows=1, names=['%d' % i for i in range(0, N_rep)]).reset_index().drop(['index'], axis=1)
df_tindex
```

```python
df_temp0 = pd.read_table('temp0.xvg', delim_whitespace=True,  skiprows=19, names=['temp0']).reset_index()
df_temp1 = pd.read_table('temp1.xvg', delim_whitespace=True,  skiprows=19, names=['temp1']).reset_index()
df_temp2 = pd.read_table('temp2.xvg', delim_whitespace=True,  skiprows=19, names=['temp2']).reset_index()
df_temp3 = pd.read_table('temp3.xvg', delim_whitespace=True,  skiprows=19, names=['temp3']).reset_index()
df_temp4 = pd.read_table('temp4.xvg', delim_whitespace=True,  skiprows=19, names=['temp4']).reset_index()
df_temp5 = pd.read_table('temp5.xvg', delim_whitespace=True,  skiprows=19, names=['temp5']).reset_index()
df_temp6 = pd.read_table('temp6.xvg', delim_whitespace=True,  skiprows=19, names=['temp6']).reset_index()
df_temp7 = pd.read_table('temp7.xvg', delim_whitespace=True,  skiprows=19, names=['temp7']).reset_index()
df_temp8 = pd.read_table('temp8.xvg', delim_whitespace=True,  skiprows=19, names=['temp8']).reset_index()
df_temp9 = pd.read_table('temp9.xvg', delim_whitespace=True,  skiprows=19, names=['temp9']).reset_index()
df_temp10 = pd.read_table('temp10.xvg', delim_whitespace=True,  skiprows=19, names=['temp10']).reset_index()
df_temp11 = pd.read_table('temp11.xvg', delim_whitespace=True,  skiprows=19, names=['temp11']).reset_index()
df_temp = pd.concat([df_temp0, df_temp1, df_temp2, df_temp3, df_temp4, df_temp5, \
                                 df_temp6, df_temp7, df_temp8, df_temp9, df_temp10, df_temp11], axis=1).reset_index().drop(['index', 'level_0'], axis=1)
df_temp
```

```python
# Plot
temp_list = [[] for i in range(0, N_rep)]
index = 1
addind = 3

for i, v in df_temp.iterrows():
    for j in range(0, N_rep):
        if i == 0:
            temp_list[j].append(v[j])
        else:
                tindex = df_tindex.iloc[index, j]
                temp_list[j].append(v[tindex])

    if i != 0:
        index += addind
        if addind == 3:
            addind = 2
        else:
            addind = 3
```

```python
sns.set_style('ticks')
mpl.rc('font', weight='bold')
mpl.rcParams['axes.linewidth'] = 2.5

plt.plot(temp_list[0], linewidth=0.4, color='Red')
```
