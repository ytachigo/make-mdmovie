```python
# Color gradation
gain = 10
offset_x= 0.2
offset_green = 0.6

def sigmoid(x, gain=1, offset_x=0):
    return ((np.tanh(((x + offset_x) * gain) / 2) + 1) / 2)

def colorbar_rgb(x):
    x = (x * 2) - 1
    red = sigmoid(x, gain, - 1 * offset_x)
    blue = 1 - sigmoid(x, gain, offset_x)
    green = sigmoid(x, gain, offset_green) + (1 - sigmoid(x, gain, - 1 * offset_green))
    green = green - 1.0
    return (blue * 255, green * 255, red * 255) # 8 bit (0 ~ 255)
```

```python
# Make a movie
N_step = int(len(df_avetemp) / 4)
N_para = 12
temp_max = df_avetemp.iloc[:, 2].max()
temp_min = df_avetemp.iloc[:, 2].min()
aveco_max = df_avetemp.iloc[:, 1].max()
aveco_min = df_avetemp.iloc[:, 1].min()

for i in range(0, N_step + 1):
    read_file = 'figures/untitled.{0:05d}.ppm'.format(i)
    write_file = 'figures/write.{0:05d}.ppm'.format(i)
    temp = 'Temperature: {:.5} K'.format(df_avetemp.iloc[i * 4, 2])
    aveco = 'The average coordination number: {:.3}'.format(df_avetemp.iloc[i * 4, 1] / N_para)
    img = cv.imread(read_file)
    cv.putText(img, temp, (5, 30), cv.FONT_HERSHEY_SIMPLEX, 0.8, \
                    colorbar_rgb((df_avetemp.iloc[i * 4, 2] - temp_min) / (temp_max - temp_min)), 2)
    cv.putText(img, aveco, (5, 65), cv.FONT_HERSHEY_SIMPLEX, 0.8, \
                    colorbar_rgb((df_avetemp.iloc[i * 4, 1] - aveco_min) / (aveco_max - aveco_min)), 2)
    cv.imwrite(write_file, img)
```
